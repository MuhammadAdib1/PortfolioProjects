```python
# Import data frames from January - December

jan_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202101-divvy-tripdata.csv")
feb_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202102-divvy-tripdata.csv")
mac_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202103-divvy-tripdata.csv")
apr_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202104-divvy-tripdata.csv")
may_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202105-divvy-tripdata.csv")
jun_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202106-divvy-tripdata.csv")
jul_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202107-divvy-tripdata.csv")
aug_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202108-divvy-tripdata.csv")
sep_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202109-divvy-tripdata.csv")
oct_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202110-divvy-tripdata.csv")
nov_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202111-divvy-tripdata.csv")
dec_2021 = pd.read_csv (r"C:\Users\madib\Desktop\Case Study 1\202112-divvy-tripdata.csv")
```

```python
#Merge the data frames becoming a dataset

dataset = [jan_2021,feb_2021,mac_2021,apr_2021,may_2021,jun_2021,aug_2021,sep_2021,oct_2021,nov_2021,dec_2021]
dataset = pd.concat(dataset)
```

```python
#Inspecting at the dataset info 

dataset.info()
```

```python
#Looking for null values within the dataset

dataset.isnull().sum()
```

```python
#Remove the nulls from the dataset

dataset = dataset.dropna(how='any', axis=0)

print('Null values remove successfully')

dataset.isnull().sum()
```

```python
#Looking for any duplicates row

dataset.duplicated().any()
```

```python
#Change the data types

pd.options.mode.chained_assignment = None

dataset["started_at"] = pd.to_datetime(dataset["started_at"])
dataset["ended_at"] = pd.to_datetime(dataset["ended_at"])

dataset.info()
```

```python
#Creating new metrics for better insights from dataset

dataset["trip_duration"] = dataset["ended_at"] - dataset["started_at"]
dataset["trip_duration"] = pd.to_numeric(dataset["trip_duration"])/6e+10
```

```python
dataset["trip_duration"].describe(include = "all")
```

```python
# removing outliers by creating filters

outliers_trip_duration = (dataset["trip_duration"] < 1) | (dataset["trip_duration"] > 480)
outliers_trip_duration.value_counts()
```

```python
# delete conflicting values

dataset.drop(dataset.index[outliers_trip_duration], inplace = True)
```

```python
# getting the corresponding linear distances in meters using Numpy Library function

dataset["trip_distance"] = np.sqrt(((dataset["end_lat"] - dataset["start_lat"])**2) + ((dataset["end_lng"] - dataset["start_lng"])**2))
dataset["trip_distance"] = dataset["trip_distance"] * 111.139
```

```python
dataset["start_hour"] = dataset["started_at"].dt.hour
dataset["start_hour"].unique()
```

```python
dataset["start_day_num"] = dataset["started_at"].dt.dayofweek
day_dict = {0:"Sunday", 1:"Monday", 2:"Tuesday", 3:"Wednesday", 4:"Thursday", 5:"Friday", 6:"Saturday"}
dataset["start_day_name"]= dataset["start_day_num"].apply(lambda y:day_dict[y])
```

```python
dataset["weekend"] = np.where(dataset["start_day_num"] > 5, True, False)
```

```python
# adding more columns such as start_month and start_year from datetime library

dataset['start_month_num'] = dataset['started_at'].dt.month
dataset["start_month_num"].unique()
```

```python
month_dict = {1:"January", 2:"February", 3:"March", 4:"April", 5:"May", 6:"June", 7:"July", 8:"August", 9:"September", 10:"October", 11:"November", 12:"December"}
dataset["start_month_name"]= dataset["start_month_num"].apply(lambda y:month_dict[y])
dataset["start_month_name"].unique()
```

```python
dataset['start_year'] = dataset['started_at'].dt.year
dataset["start_year"].unique()
```

```python
dataset.rename(columns = {'member_casual': 'user_type'}, inplace=True)
```
